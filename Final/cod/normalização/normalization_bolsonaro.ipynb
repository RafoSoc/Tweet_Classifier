```python
# Importando dataset
dataset = pd.read_csv('C:/Users/rafae/OneDrive/Documents/GitHub/HCAI/Final/dados/raw_data/bolsonaro3.csv', 
                        parse_dates=['created_at'], dtype={'id_str':str})
display(dataset.head())
```

```python
# Substitui os valores NAN da coluna retweeted_status.full_text pelos valores correspondentes da linha da coluna full_text
dataset['all_full_text'] = dataset['retweeted_status.full_text'].fillna(dataset['full_text'])
display(dataset)
```
